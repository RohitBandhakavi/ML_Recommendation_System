If you see theat thre are multiple with past present and future sence like love, loved, loving.  so to club all those words into one we use stemming which will make all those words as one like love. For this we use nltk library which is nltk.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()   # we do this for ease of use.
```

```python
def stem(text):
    y = []

    for i in text.split():    # doing this to convert the string to list so that each word can be stemmed.
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
# Example of stemming

ps.stem('loved')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

Now for the recommendation process we need to get the similarities bw tags so that id a person likes mission impossible then we can recommend him similar movies that is the most similar tags to that.
for that we are converting the string('tags') to vectors so that we can be able to plot them on the graph.
we are using text vectorization technique called bag of words.

In the bag of words model, we combine all the tags and lets say we have 5L words. then we take the n number od most used words(lets say n = 5000). now our thing is to creat a DF with all these most frequent vectors and the score of the appreance of these vectors in the tags of every single movie.
if we plot these scores in the graph then we can see that theres a movie A and then we can see the 5 nearest movies to that and recommend them to the user.
that means here every movie will become a vector in a n-dimensional space.
we are ot considering any stop words here cuz they dont make any sence to consider them.

```python
# importing scikit learn module for the vectorization.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() 
```

We use toarray cuz normally transform from sklearn returns a sparse matrix which only stores the position and the value of the non zero numbers in the matrix. So if we convert it to a numpy array(which is what.toarray does), then we store the whole matrix insted of the nonzero values.

```python
np.set_printoptions(threshold=np.inf)  # this will make us see the full output.
cv.get_feature_names_out()
```

In this program, we use cosine distance to measure similarity between 5000-dimensional movie embeddings. Cosine distance focuses on the angle between vectors, which reflects how similar their directions are, regardless of their magnitude. This is important because two movies can have similar themes or genres (i.e., similar direction) even if their feature magnitudes differ. Euclidean distance, on the other hand, is sensitive to magnitude and performs poorly in high-dimensional spaces due to the curse of dimensionality. Therefore, cosine distance provides more meaningful and reliable results for content-based movie recommendations.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

here the similarity[0] means the similarity of the movie 1 with the rest of the 4806 movies. and since the movie 1 is exactly same to the first movie, so thats why the similarity score is 1.

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  #fetching the index of the movie that is given in the function
    distance = similarity[movie_index]   # this fetched the similarity score of the movie with the index.
    movie_list = sorted(list(enumerate(distance)), reverse= True, key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
  
```

This code finds and prints the top 5 movies most similar to a given movie based on a precomputed similarity matrix. First, it gets the index of the movie from the DataFrame new by matching the movie title

Then it retrieves the similarity scores between this movie and all others, pairs each score with its index using enumerate, and sorts them in descending order based on the similarity value:

Finally, it loops through the top 5 most similar movies (excluding the original movie, which is always most similar to itself) and prints their titles using their indices from the sorted list

key = lambda x: x[1] tells the sorted() function to sort based on the second element of each tuple. the enumerate op will be (movie indec, similarity score)
In this case, each item is a tuple like (movie_index, similarity_score), so x[1] is the similarity score.
This ensures the sorting is done by similarity values, not by the index.



```python
recommend('Avatar') # these are the recommended movies.
```
